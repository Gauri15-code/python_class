```python
import tkinter as tk
import math

# Q1
root = tk.Tk()
root.title("Robot Control Panel")
root.geometry("500x400")
root.configure(bg='yellow')
root.mainloop()

# Q2
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400)
canvas.pack()
canvas.create_oval(100, 100, 102, 102, fill='black')
root.mainloop()

# Q3
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400)
canvas.pack()
canvas.create_line(20,50, 100,120, 180,90, 250,150, fill='blue', width=3)
root.mainloop()

# Q4
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400)
canvas.pack()
point = canvas.create_oval(0, 195, 10, 205, fill='red')
def move():
    canvas.move(point, 1, 0)
    if canvas.coords(point)[0] < 500:
        root.after(10, move)
move()
root.mainloop()

# Q5
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400)
canvas.pack()
canvas.create_rectangle(200, 150, 300, 250, fill='gray')
canvas.create_oval(210, 250, 240, 280, fill='black')
canvas.create_oval(260, 250, 290, 280, fill='black')
root.mainloop()

# Q6
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400, bg='white')
canvas.pack()
robot = canvas.create_oval(245, 195, 255, 205, fill='blue')
def move(dx, dy):
    canvas.move(robot, dx, dy)
tk.Button(root, text='Up', command=lambda: move(0, -10)).pack(side=tk.LEFT)
tk.Button(root, text='Down', command=lambda: move(0, 10)).pack(side=tk.LEFT)
tk.Button(root, text='Left', command=lambda: move(-10, 0)).pack(side=tk.LEFT)
tk.Button(root, text='Right', command=lambda: move(10, 0)).pack(side=tk.LEFT)
root.mainloop()

# Q7
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400, bg='white')
canvas.pack()
ball = canvas.create_oval(185, 135, 215, 165, fill='red')
vx, vy = 2, 2
def bounce():
    global vx, vy
    x1, y1, x2, y2 = canvas.coords(ball)
    if x1 <= 0 or x2 >= 500:
        vx = -vx
    if y1 <= 0 or y2 >= 400:
        vy = -vy
    canvas.move(ball, vx, vy)
    root.after(10, bounce)
bounce()
root.mainloop()

# Q8
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400, bg='white')
canvas.pack()
robot = canvas.create_oval(50, 200, 60, 210, fill='green')
x = 50
def move_to_target():
    global x
    if x < 450:
        x += 2
        canvas.coords(robot, x, 200, x+10, 210)
        root.after(20, move_to_target)
move_to_target()
root.mainloop()

# Q9
root = tk.Tk()
canvas = tk.Canvas(root, width=600, height=500, bg='white')
canvas.pack()
L1, L2, L3, L4 = 250, 120, 150, 130
A = (150, 300)
D = (400, 300)
theta = math.radians(30)
B = (A[0] + L2*math.cos(theta), A[1] + L2*math.sin(theta))
d = math.hypot(D[0]-B[0], D[1]-B[1])
a = (L3**2 - L4**2 + d**2) / (2*d)
h = math.sqrt(L3**2 - a**2)
x2 = B[0] + a*(D[0]-B[0])/d
y2 = B[1] + a*(D[1]-B[1])/d
x3 = x2 - h*(D[1]-B[1])/d
y3 = y2 + h*(D[0]-B[0])/d
C = (x3, y3)
canvas.create_line(A, B, fill='red', width=3)
canvas.create_line(B, C, fill='green', width=3)
canvas.create_line(C, D, fill='blue', width=3)
canvas.create_line(A, D, fill='black', width=3)
canvas.create_oval(A[0]-5, A[1]-5, A[0]+5, A[1]+5, fill='black')
canvas.create_oval(B[0]-5, B[1]-5, B[0]+5, B[1]+5, fill='black')
canvas.create_oval(C[0]-5, C[1]-5, C[0]+5, C[1]+5, fill='black')
canvas.create_oval(D[0]-5, D[1]-5, D[0]+5, D[1]+5, fill='black')
root.mainloop()

# Q10
root = tk.Tk()
canvas = tk.Canvas(root, width=500, height=400, bg='white')
canvas.pack()
robot = canvas.create_oval(245, 195, 255, 205, fill='blue')
path = []
def move_robot(event):
    x1, y1, x2, y2 = canvas.coords(robot)
    cx, cy = (x1+x2)/2, (y1+y2)/2
    if event.keysym == 'Up':
        canvas.move(robot, 0, -5)
    elif event.keysym == 'Down':
        canvas.move(robot, 0, 5)
    elif event.keysym == 'Left':
        canvas.move(robot, -5, 0)
    elif event.keysym == 'Right':
        canvas.move(robot, 5, 0)
    nx1, ny1, nx2, ny2 = canvas.coords(robot)
    ncx, ncy = (nx1+nx2)/2, (ny1+ny2)/2
    path.append((cx, cy, ncx, ncy))
    canvas.create_line(cx, cy, ncx, ncy, fill='gray')
def reset():
    canvas.delete('all')
    global robot, path
    robot = canvas.create_oval(245, 195, 255, 205, fill='blue')
    path = []
canvas.bind_all('<KeyPress>', move_robot)
tk.Button(root, text='RESET', command=reset).pack()
root.mainloop()
```
