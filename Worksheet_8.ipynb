```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def midpoint(self, other):
        return Point((self.x + other.x)/2, (self.y + other.y)/2)

    def line_equation(self, other):
        if self.x == other.x:
            return f"x = {self.x}"
        m = (other.y - self.y) / (other.x - self.x)
        c = self.y - m * self.x
        return f"y = {m:.2f}x + {c:.2f}"

    def reflection_over_line(self, other, point):
        if self.x == other.x:
            return Point(2*self.x - point.x, point.y)
        m = (other.y - self.y) / (other.x - self.x)
        c = self.y - m * self.x
        d = (point.x + (point.y - c)*m) / (1 + m**2)
        x_ref = 2*d - point.x
        y_ref = 2*d*m - point.y + 2*c
        return Point(x_ref, y_ref)

# Question 1
x1, y1 = 1, 2
x2, y2 = 4, 6
x3, y3 = 2, 3
A = Point(x1, y1)
B = Point(x2, y2)
C = Point(x3, y3)

print("Distance:", A.distance(B))
mid = A.midpoint(B)
print("Midpoint:", (mid.x, mid.y))
print("Line equation:", A.line_equation(B))
reflection = A.reflection_over_line(B, C)
print("Reflection of C:", (reflection.x, reflection.y))

# Question 2
A_vec = [3, 4]
B_vec = [1, 2]
C_vec = [5, 1]

R = [A_vec[0]+B_vec[0]+C_vec[0], A_vec[1]+B_vec[1]+C_vec[1]]
print("\nResultant vector:", R)

def magnitude(v):
    return math.sqrt(v[0]**2 + v[1]**2)
print("Magnitude A:", magnitude(A_vec))
print("Magnitude B:", magnitude(B_vec))
print("Magnitude C:", magnitude(C_vec))

def dot(v1, v2):
    return v1[0]*v2[0] + v1[1]*v2[1]
print("A·B:", dot(A_vec, B_vec))
print("A·C:", dot(A_vec, C_vec))
print("B·C:", dot(B_vec, C_vec))

def angle(v1, v2):
    cos = dot(v1, v2) / (magnitude(v1) * magnitude(v2))
    return math.degrees(math.acos(cos))
print("Angle A-B:", angle(A_vec, B_vec))
print("Angle A-C:", angle(A_vec, C_vec))
print("Angle B-C:", angle(B_vec, C_vec))

proj = (dot(A_vec, B_vec) / magnitude(B_vec)**2)
projection = [proj * B_vec[0], proj * B_vec[1]]
print("Projection of A on B:", projection)

# Question 3
S = Point(1, 1)
E = Point(4, 5)
P = Point(2, 2)

seg_length = S.distance(E)
print("\nSegment length:", seg_length)

def closest_point_on_segment(S, E, P):
    sx, sy = S.x, S.y
    ex, ey = E.x, E.y
    px, py = P.x, P.y

    dx = ex - sx
    dy = ey - sy
    t = ((px - sx)*dx + (py - sy)*dy) / (dx*dx + dy*dy)
    t = max(0, min(1, t))
    closest_x = sx + t*dx
    closest_y = sy + t*dy
    return Point(closest_x, closest_y)

closest = closest_point_on_segment(S, E, P)
print("Closest point on segment:", (closest.x, closest.y))
print("Distance from P to segment:", P.distance(closest))

# Q4
def line_intersection(a1, b1, c1, a2, b2, c2):
    det = a1*b2 - a2*b1
    if det == 0:
        return "Lines are parallel or coincident."
    x = (c1*b2 - c2*b1) / det
    y = (a1*c2 - a2*c1) / det
    return (x, y)

print(line_intersection(2, 3, 5, 4, 1, 3))
print(line_intersection(1, 2, 3, 2, 4, 6))
print(line_intersection(1, 1, 5, 2, 2, 10))
```
