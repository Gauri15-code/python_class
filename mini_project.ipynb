```python
import numpy as np

# Project 1: Sensor Data Analysis
sensor_data = np.genfromtxt('sensor_data.csv', delimiter=',')
temperatures = sensor_data[:, 1]
distances = sensor_data[:, 2]
batteries = sensor_data[:, 3]

avg_temp = np.mean(temperatures)
min_temp = np.min(temperatures)
max_temp = np.max(temperatures)
avg_dist = np.mean(distances)
min_dist = np.min(distances)
max_dist = np.max(distances)
avg_batt = np.mean(batteries)
min_batt = np.min(batteries)
max_batt = np.max(batteries)

time_column = sensor_data[:, 0]
time_of_max_temp = time_column[np.argmax(temperatures)]
low_battery_count = np.sum(batteries < 30)

print("Sensor Data Analysis:")
print("Average Temperature: {:.2f}°C, Min: {:.2f}°C, Max: {:.2f}°C".format(avg_temp, min_temp, max_temp))
print("Average Distance: {:.2f}m, Min: {:.2f}m, Max: {:.2f}m".format(avg_dist, min_dist, max_dist))
print("Average Battery: {:.2f}%, Min: {:.2f}%, Max: {:.2f}%".format(avg_batt, min_batt, max_batt))
print("Time when temperature was highest: {} minutes".format(time_of_max_temp))
print("Number of times battery was below 30%: {}".format(low_battery_count))

with open('sensor_results.txt', 'w') as f:
    f.write("Sensor Data Analysis Results\n")
    f.write("----------------------------\n")
    f.write("Average Temperature: {:.2f}°C, Min: {:.2f}°C, Max: {:.2f}°C\n".format(avg_temp, min_temp, max_temp))
    f.write("Average Distance: {:.2f}m, Min: {:.2f}m, Max: {:.2f}m\n".format(avg_dist, min_dist, max_dist))
    f.write("Average Battery: {:.2f}%, Min: {:.2f}%, Max: {:.2f}%\n".format(avg_batt, min_batt, max_batt))
    f.write("Time when temperature was highest: {} minutes\n".format(time_of_max_temp))
    f.write("Number of times battery was below 30%: {}\n".format(low_battery_count))

print("\nResults saved to sensor_results.txt.")

print("\n-----------------------------------------\n")

# Project 2: Robot Path Analysis
robot_path = np.genfromtxt('robot_path.csv', delimiter=',')
diffs = np.diff(robot_path, axis=0)
distances = np.sqrt(np.sum(diffs**2, axis=1))
total_distance = np.sum(distances)

distances_from_origin = np.sqrt(np.sum(robot_path**2, axis=1))
farthest_index = np.argmax(distances_from_origin)
farthest_point = robot_path[farthest_index]

unique_points, counts = np.unique(robot_path, axis=0, return_counts=True)
revisited = unique_points[counts > 1]

print("\nRobot Path Analysis:")
print("Total distance traveled: {:.2f} units".format(total_distance))
print("Farthest point from origin: {}".format(farthest_point))
if len(revisited) > 0:
    print("Revisited positions: {}".format(revisited))
else:
    print("The robot never revisited a position.")

with open('path_results.txt', 'w') as f:
    f.write("Robot Path Analysis Results\n")
    f.write("---------------------------\n")
    f.write("Total distance traveled: {:.2f} units\n".format(total_distance))
    f.write("Farthest point from origin: {}\n".format(farthest_point))
    if len(revisited) > 0:
        f.write("Revisited positions: {}\n".format(revisited))
    else:
        f.write("The robot never revisited a position.\n")

print("Results saved to path_results.txt.")
```
