```python
import pandas as pd
print(pd.__version__)

data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)

S1 = pd.Series([100, 200, 300, 400, 500])
print(S1)
print(S1[1], S1[3])

S2 = pd.Series([10, 20, 30, 40, 50])
print(S1 + S2)

print(df[['Name', 'City']])

df['Salary'] = [50000, 60000, 70000]
print(df)

print(df['Age'].mean())
print(df['Salary'].sum())

print(df[df['Age'] > 28])

df = df.set_index('Name')
print(df)

df = df.reset_index()
print(df)

import pandas as pd

employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Emily'],
    'Department': ['Sales', 'Marketing', 'HR'],
    'Salary': [50000, 60000, 55000]
})

print(employees)

filtered = employees[employees['Salary'] > 55000]
print(filtered[['Name', 'Department']])

grouped = employees.groupby('Department')['Salary'].mean()
print(grouped)

agg_result = employees.groupby('Department')['Salary'].agg(['min', 'max'])
print(agg_result)

df1 = pd.DataFrame({
    'Name': ['John', 'Jane', 'Emily'],
    'Department': ['Sales', 'Marketing', 'HR']
})
df2 = pd.DataFrame({
    'Name': ['John', 'Jane', 'Emily'],
    'Experience (Years)': [5, 7, 3]
})
merged = pd.merge(df1, df2, on='Name')
print(merged)

sorted_df = merged.sort_values(by='Experience (Years)', ascending=False)
print(sorted_df)
```
