```python
import numpy as np
from scipy import stats, fftpack, linalg, interpolate, signal
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Q1
np.random.seed(42)
arr = np.random.rand(100)
mean = np.mean(arr)
median = np.median(arr)
variance = np.var(arr)

# Q2
arr2d = np.random.rand(5, 5)
fft_result = fftpack.fft2(arr2d)

# Q3
matrix = np.random.rand(3, 3)
determinant = linalg.det(matrix)
inverse = linalg.inv(matrix)
eigenvalues = linalg.eigvals(matrix)

# Q4
x = np.linspace(0, 10, 10)
y = np.sin(x)
f = interpolate.interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 50)
y_new = f(x_new)

# Q5
t = np.linspace(0, 1, 1000)
signal_data = np.sin(2*np.pi*5*t) + 0.5*np.random.randn(1000)
b, a = signal.butter(4, 0.1)
filtered_signal = signal.filtfilt(b, a, signal_data)

# Q6
np.random.seed(42)
sales = np.random.randint(100, 500, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
products = ['Product_A', 'Product_B', 'Product_C', 'Product_D']

total_sales = np.sum(sales)
average_sales = np.mean(sales)
max_sales = np.max(sales)
monthly_totals = np.sum(sales, axis=1)
best_month_index = np.argmax(monthly_totals)
worst_month_index = np.argmin(monthly_totals)
best_month = months[best_month_index]
worst_month = months[worst_month_index]
product_totals = np.sum(sales, axis=0)
best_product_index = np.argmax(product_totals)
worst_product_index = np.argmin(product_totals)
best_product = products[best_product_index]
worst_product = products[worst_product_index]

# Q7 + Q9
names = ['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal']
math = np.array([85, 79, 90, 66, 70])
physics = np.array([78, 82, 85, 75, 68])
chemistry = np.array([92, 74, 89, 80, 75])
english = np.array([88, 90, 92, 78, 85])
marks = np.column_stack((math, physics, chemistry, english))

total_marks = np.sum(marks, axis=1)
average_marks = np.mean(marks, axis=1)
subject_means = np.mean(marks, axis=0)
subject_max = np.max(marks, axis=0)
subject_min = np.min(marks, axis=0)

top_performer_index = np.argmax(total_marks)
bottom_performer_index = np.argmin(total_marks)
top_performer = names[top_performer_index]
bottom_performer = names[bottom_performer_index]

passing_threshold = 40
passed = np.sum(marks >= passing_threshold, axis=0)
total_students = marks.shape[0]
passing_percentage = (passed / total_students) * 100

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(names, total_marks, color='skyblue')
plt.title('Total Marks per Student')
plt.ylabel('Total Marks')

plt.subplot(1, 2, 2)
subjects = ['Math', 'Physics', 'Chemistry', 'English']
plt.bar(subjects, subject_means, color='lightgreen')
plt.title('Average Marks per Subject')
plt.ylabel('Average Marks')

plt.tight_layout()
plt.show()

# Q8
time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, covariance = curve_fit(quadratic, time, velocity)
a_fit, b_fit, c_fit = params

t_fit = np.linspace(0, 5, 100)
v_fit = quadratic(t_fit, a_fit, b_fit, c_fit)

plt.figure()
plt.scatter(time, velocity, color='red', label='Data')
plt.plot(t_fit, v_fit, label=f'Fit: v(t)={a_fit:.2f}t+{b_fit:.2f}t+{c_fit:.2f}')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.title('Velocity Curve Fitting')
plt.show()

# Q10
time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, _ = curve_fit(quadratic, time, velocity)
a_fit, b_fit, c_fit = params

t_fit = np.linspace(0, 5, 100)
v_fit = quadratic(t_fit, a_fit, b_fit, c_fit)

plt.figure(figsize=(8, 6))
plt.scatter(time, velocity, color='red', s=80, label='Original Data', zorder=5)
plt.plot(t_fit, v_fit, linewidth=2, label=f'Fit: v(t)={a_fit:.2f}t+{b_fit:.2f}t+{c_fit:.2f}')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Velocity (m/s)', fontsize=12)
plt.title('Velocity Curve Fitting', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# Q11
years = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([50, 55, 70, 80, 90])

correlation, _ = stats.pearsonr(years, population)
interp_func = interpolate.interp1d(years, population, kind='linear')
pop_2008 = interp_func(2008)

plt.figure(figsize=(8, 6))
plt.scatter(years, population, color='blue', s=80, label='Data Points', zorder=5)
plt.plot(years, population, 'g--', alpha=0.7, label='Linear Interpolation')
plt.scatter(2008, pop_2008, color='red', s=120, zorder=6, label=f'2008 Estimate: {pop_2008:.1f}k')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (thousands)', fontsize=12)
plt.title('Population Interpolation', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# Q12
coeff = [3, -5, 2, -8]
roots = np.roots(coeff)

x = np.linspace(-3, 3, 400)
y = 3*x**3 - 5*x**2 + 2*x - 8

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2, label='p(x)=3x-5x+2x-8')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.scatter(roots, np.zeros_like(roots), color='red', s=100, zorder=5, label='Roots')
plt.xlabel('x', fontsize=12)
plt.ylabel('p(x)', fontsize=12)
plt.title('Polynomial with Roots', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# Q13
sizes = [200, 400, 600, 800, 1000]
times = [0.8, 1.5, 2.3, 3.1, 3.9]

plt.plot(sizes, times, 'bo-')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.title('File Processing Time')
plt.grid(True)
plt.show()

# Q14
def f(x):
    return x**4 - 3*x**3 + 2

x = np.linspace(-2, 3, 400)
y = f(x)
min_index = np.argmin(y)
min_x = x[min_index]
min_y = y[min_index]

plt.plot(x, y, 'b-')
plt.scatter(min_x, min_y, color='red', s=100)
plt.axhline(y=0, color='k', alpha=0.3)
plt.axvline(x=0, color='k', alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function with Local Minima')
plt.grid(True)
plt.show()

# Q15
t = np.linspace(0, 20, 400)
theta = np.exp(-0.1*t) * np.cos(t)

plt.plot(t, theta, 'b-')
plt.xlabel('Time (s)')
plt.ylabel('(t)')
plt.title('Damped Oscillatory System')
plt.grid(True)
plt.show()
```
